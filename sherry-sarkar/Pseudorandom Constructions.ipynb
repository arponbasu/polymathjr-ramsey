## Implementation of the Exercise 5 Construction 

```python
x = vector(GF(2), [0, 1, 1, 0])
y = vector(GF(2), [1, 1, 0, 1])
```

```python
# Takes in two 2n dimensional vectors over F_2, and outputs a 0 or 1. 
def beta(u, v):
    
    if len(u) % 2 != 0 or len(v) % 2 != 0: 
        raise Exception("Input vectors need to have even dimension.")
    
    i = 0 
    B = 0
    while i < len(u)/2: 
        B += u[i] * v[i + 1] - u[i + 1] * v[i]
        i += 2
    
    return B
```

```python
beta(x, y)
```

```python
V = VectorSpace(GF(2), 8)
vectors = list(V)
vectors.remove(vector(GF(2), [0,0,0,0,0,0,0,0]))
```

```python
annoying_vectors = [tuple(vec) for vec in vectors]
```

```python
construction = Graph([annoying_vectors, lambda u, v: beta(u, v) == 1])
```

```python
construction.random_subgraph(0.05).plot()
```
