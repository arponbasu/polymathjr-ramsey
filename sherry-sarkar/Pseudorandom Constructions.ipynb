## Implementation of the Exercise 5 Construction 

```python
x = vector(GF(2), [1, 0, 1, 0], immutable=True)
y = vector(GF(2), [1, 1, 0, 1])
```

```python
# Takes in two 2n dimensional vectors over F_2, and outputs a 0 or 1. 
def beta(u, v):
    
    if len(u) != len(v) or len(u) % 2 != 0: 
        raise Exception("Input vectors need to have equal and even dimension.")
    
    B = 0
    for i in range(0, len(u), 2): 
        B += u[i] * v[i + 1] - u[i + 1] * v[i]
    
    return B
```

```python
beta(x, y)
```

```python
V = VectorSpace(GF(2), 8)
vectors = list(V)
vectors.remove(vector(GF(2), [0,0,0,0,0,0,0,0]))
for vec in vectors: vec.set_immutable()
```

```python
construction = Graph([vectors, lambda u, v: beta(u, v) == 1])
```

```python
construction.random_subgraph(0.05).plot()
```

```python
paley17=graphs.PaleyGraph(17)
```

```python
subg = construction.subgraph_search(paley17, induced=True)
```

```python
subg.show()
print(subg.vertices())
print(subg.clique_number(), subg.complement().clique_number())
```
